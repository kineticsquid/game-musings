## Analyze Game Results

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
import compute_scores
import analyze_data
```

```python
all_input = analyze_data.get_input()
results = analyze_data.process_input(all_input)
total_hands = results['total_hands']
total_second_round_hands = results['total_second_round_hands']
nines_and_tens = results['nines_and_tens']
farmers_hands = results['farmers_hands']
player_first_round_hand_scores = results['player_first_round_hand_scores']
first_round_team_scores = results['first_round_team_scores']
player_second_round_hand_scores = results['player_second_round_hand_scores']
second_round_team_scores = results['second_round_team_scores']
total_trump_calls = results['total_trump_calls']
trump_called_hand_scores = results['trump_called_hand_scores']
```

```python
plt.rcParams.update({'font.size': 16})
fig = plt.figure()
fig = plt.figure(figsize=(7.0,5.0))
ax = fig.add_axes([0,0,1,1])
ax.set_title('Average First Round Hand Scores (%s total hands)' % total_hands)
bar_values = [i/total_hands for i in list(player_first_round_hand_scores.values())]
ax.bar(list(player_first_round_hand_scores.keys()), bar_values)
for i in range(0, len(list(player_first_round_hand_scores.values()))):
    plt.text(i, bar_values[i] - 0.3, '%1.1f' % bar_values[i], ha='center')
plt.show()
```

```python
plt.rcParams.update({'font.size': 16})
fig = plt.figure()
fig = plt.figure(figsize=(7.0,5.0))
ax = fig.add_axes([0,0,1,1])
ax.set_title('Average First Round Team Scores (%s total hands)' % total_hands)
bar_values = [i/total_hands for i in list(first_round_team_scores.values())]
ax.bar(list(first_round_team_scores.keys()), bar_values)
for i in range(0, len(list(first_round_team_scores.values()))):
    plt.text(i, bar_values[i] - 0.7, '%1.1f' % bar_values[i], ha='center')
plt.show()
```

```python
plt.rcParams.update({'font.size': 16})
fig = plt.figure()
fig = plt.figure(figsize=(7.0,5.0))
ax = fig.add_axes([0,0,1,1])
ax.set_title('Average Second Round Hand Scores (%s total hands)' % total_second_round_hands)
bar_values = [i/total_second_round_hands for i in list(player_second_round_hand_scores.values())]
ax.bar(list(player_second_round_hand_scores.keys()), bar_values)
for i in range(0, len(list(player_second_round_hand_scores.values()))):
    plt.text(i, bar_values[i] - 0.3, '%1.1f' % bar_values[i], ha='center')
plt.show()
```

```python
plt.rcParams.update({'font.size': 16})
fig = plt.figure()
fig = plt.figure(figsize=(7.0,5.0))
ax = fig.add_axes([0,0,1,1])
ax.set_title('Average Second Round Team Scores (%s total hands)' % total_second_round_hands)
bar_values = [i/total_second_round_hands for i in list(second_round_team_scores.values())]
ax.bar(list(second_round_team_scores.keys()), bar_values)
for i in range(0, len(list(second_round_team_scores.values()))):
    plt.text(i, bar_values[i] - 0.7, '%1.1f' % bar_values[i], ha='center')
plt.show()
```

```python
plt.rcParams.update({'font.size': 16})
fig = plt.figure()
fig = plt.figure(figsize=(7.0,5.0))
ax = fig.add_axes([0,0,1,1])
ax.set_title('Number of Trump Calls (%s total hands)' % total_hands)
ax.bar(list(total_trump_calls.keys()), list(total_trump_calls.values()))
for i in range(0, len(list(total_trump_calls.values()))):
    plt.text(i, list(total_trump_calls.values())[i] - 4, list(total_trump_calls.values())[i], ha='center')
plt.show()
```

```python
plt.rcParams.update({'font.size': 16})
fig = plt.figure()
fig = plt.figure(figsize=(7.0,5.0))
ax = fig.add_axes([0,0,1,1])
ax.set_title('Average Hand Score on Trump Call')
bar_values = []
for i in range(0, len(total_trump_calls)):
    bar_values.append(list(trump_called_hand_scores.values())[i]/list(total_trump_calls.values())[i])
ax.bar(list(total_trump_calls.keys()), bar_values)
for i in range(0, len(list(total_trump_calls.values()))):
    plt.text(i, bar_values[i] - 0.7, '%1.1f' % bar_values[i], ha='center')
plt.show()
```

```python
plt.rcParams.update({'font.size': 16})
fig = plt.figure()
fig = plt.figure(figsize=(7.0,5.0))
ax = fig.add_axes([0,0,1,1])
ax.set_title('Farmer\'s Hands (%s total hands)' % total_hands)
ax.bar(list(farmers_hands.keys()),list(farmers_hands.values()))
for i in range(0, len(list(farmers_hands.values()))):
    plt.text(i, list(farmers_hands.values())[i] -1.75, list(farmers_hands.values())[i], ha='center')
plt.show()
```

```python
plt.rcParams.update({'font.size': 16})
fig = plt.figure()
fig = plt.figure(figsize=(7.0,5.0))
ax = fig.add_axes([0,0,1,1])
ax.set_title('Average 9s and 10s per hand (%s total hands)' % total_hands)
bar_values = [i/total_hands for i in list(nines_and_tens.values())]
ax.bar(list(nines_and_tens.keys()), bar_values)
for i in range(0, len(list(nines_and_tens.values()))):
    plt.text(i, bar_values[i] - 0.1, '%1.1f' % bar_values[i], ha='center')
plt.show()
```
