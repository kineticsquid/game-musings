```python
import sympy
from sympy.utilities.iterables import multiset_combinations
from sympy.utilities.iterables import multiset_permutations
import numpy as np
import os

# http://www.scrabbleplayers.org/w/NASPA_Zyzzyva_Linux_Installation
# Note: From a Jupyter Notebook, file open works on just the file name because it is in the same
# directory as the Notebook. Debugging from Visual Studio this fails. Need to add the directory within
# the current working directory.
file_name = 'NASPA_CSW21.txt'
letters = np.array(['N', 'O', 'E', 'L', 'S', 'T', 'R', 'I', 'A', 'D'])

def form_word_from_set(set_of_letters):
    word = ''
    for letter in set_of_letters:
        word = word + letter
    return word

five_letter_words = set()
cwd = os.getcwd()

words_file = open(file_name, 'r')
for word in words_file:
    trimmed_word = word.strip().upper()
    if len(trimmed_word) == 5:
        five_letter_words.add(trimmed_word)
words_file.close()

five_letter_combinations = multiset_combinations(letters,5)
for combo in five_letter_combinations:
    valid_words = set()
    valid_second_words = set()
    possible_word_sets = multiset_permutations(combo)
    for possible_word_set in possible_word_sets:
        word = form_word_from_set(possible_word_set)
        if word in five_letter_words:
            valid_words.add(word)
    if len(valid_words) > 0:
        remaining_letters = set(letters).difference(combo)
        possible_word_sets = multiset_permutations(remaining_letters)
        for possible_word_set in possible_word_sets:
            word = form_word_from_set(possible_word_set)
            if word in five_letter_words:
                valid_second_words.add(word)
    if len(valid_words) > 0 and len(valid_second_words) > 0:
        print("%s, %s" % (valid_words, valid_second_words))
```
